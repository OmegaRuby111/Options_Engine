IMPORTING LIBRARIES

```python
import numpy as np
from scipy.stats import norm
```

```python
#defining variables
r=float(input("Enter rate of interest="))
S=float(input("Enter Stock Price="))
K=float(input("Enter Strike Price="))
days=int(input("Enter number of days="))
T=days/365
sigma=float(input("Enter volatility="))
```

BLACKSCHOLES FUNCTION

```python
def black_scholes(r,S,K,T,sigma,option_type):
    "Calculate BS Option price for a call/put option"
    d1=(np.log(S/K)+(r+sigma**2/2)*T)/(sigma*np.sqrt(T))
    d2=d1-sigma*np.sqrt(T)
    
    try:
        if option_type=="C":
            price=S*norm.cdf(d1,0,1)-K*np.exp(-r*T)*norm.cdf(d2,0,1)
        elif option_type=="P":
            price=K*np.exp(-r*T)*norm.cdf(-d2,0,1)-S*norm.cdf(-d1,0,1)
        return price
    except:
        print("Error! Please confirm all option parameters again!")
```

RETURNING THE OPTION PRICE

```python
print("Call Option Price is= ",round(black_scholes(r,S,K,T,sigma,"C"),2))
print("Put Option Price is= ",round(black_scholes(r,S,K,T,sigma,"P"),2))
```

CALCULATION OF OPTION GREEKS

```python
d1=(np.log(S/K)+(r+sigma**2/2)*T)/(sigma*np.sqrt(T))
d2=d1-sigma*np.sqrt(T)
#Delta Call
delta_call=norm.cdf(d1,0,1)
delta_put=norm.cdf(-d1,0,1)*(-1)

#Gamma for put and call
N=1/np.sqrt(np.pi*2)*np.exp(-0.5*d1**2)
gamma=(N)/(sigma*S*np.sqrt(T))

#rho
#cleaner_rho_put=-K*T*np.exp(-r*T)*norm.cdf(-d2)
#cleaner_rho_call=K*T*np.exp(-r*T)*norm.cdf(d2)
rho_call=(K*norm.cdf(d2,0,1)*T)/np.exp(r*T)
rho_put=(K*norm.cdf(-d2,0,1)*T)/(-(np.exp(r*T)))

#vega
vega=np.sqrt(T)*S*N

#theta
c1=-((0.5)/np.sqrt(T))*sigma*S
c2=(r*K)/(np.exp(r*T))
theta_call=(c1*N)-(c2*norm.cdf(d2))
theta_put=(c1*N)+(c2*norm.cdf(-d2))


print("delta call=",round(delta_call,2))
print("delta put=",round(delta_put,2))
print("gamma=",round(gamma,2))
print("rho call=",round(rho_call,2))
print("rho put=",round(rho_put,2))
print("vega=",round(vega,2))
print("theta call=",round(theta_call,2))
print("theta put=",round(theta_put,2))
```
